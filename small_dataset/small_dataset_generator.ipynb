```python

import csv
import json

with open("a.json") as f:
    json_data = json.load(f)

# Assuming json_data is a dictionary with keys as column headers and values as rows
csv_file_path = "output.csv"

keys = json_data.keys()
print("All keys:", keys)
```

```python
#Making the list of the keys
Questions=json_data['questions']
```

```python
#Storing the questions,image_id and questions_id in lists:

questions=[]
image_id=[]
question_id=[]
for i in range(0,len(Questions)):
    #Appending values to questions
    questions.append(Questions[i]['question'])
    #Appending values to image_id
    image_id.append(Questions[i]['image_id'])
    #Appending values to questions_id
    question_id.append(Questions[i]['question_id'])
    
# Specify the CSV file path
csv_file_path = "output.csv"
data_list = list(zip(questions, image_id, question_id))
# Writing to CSV
with open(csv_file_path, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header
    writer.writerow(["Question", "Image_ID", "Question_ID"])

    # Write the data
    writer.writerows(data_list)
    
```

```python
#Now we will create the excel for the answers:
with open("answers.json") as f:
    json_data = json.load(f)

# Assuming json_data is a dictionary with keys as column headers and values as rows
csv_file_path = "output_answers.csv"

keys = json_data.keys()
print("All keys:", keys)
```

```python
#Making the list of the keys
answers=json_data['annotations']
print(f'The no of answers {len(answers)}')
print(answers[22])
print(answers[20]['answers'][9])
print(answers[20].keys())
```

```python
#We will save the answers in lists:
q_T=[]
mcq=[]
an=[]
imgId=[]
ans_T=[]


for i in range(0,len(answers)):
    for key,values in answers[i].items():
        if key=='question_type':
            q_T.append(answers[i][key])
        if key=='multiple_choice_answer':
            mcq.append(answers[i][key])
        if key=='answers':
            an.append(answers[i][key])
        if key=='image_id':
            imgId.append(answers[i][key])
        if key=='answer_type':
            ans_T.append(answers[i][key])
            

#From theses list we will create the excel
        

csv_file_path = "output_answers.csv"    
    
data_list = list(zip(q_T, mcq, an,imgId,ans_T))


with open(csv_file_path, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header
    writer.writerow(["question_type", "multiple_choice_answer", "answers","image_id", "answer_type"])

    # Write the data
    writer.writerows(data_list)
    
            
            
        
        
        
        
```
